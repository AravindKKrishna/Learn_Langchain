```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="nomic-embed-text") # by default it uses llama2 
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first character of Greek alphabet",
        "Beta is the second character of Greek alphabet",
    ]
)
```

```python
r1[1]
```

```python
embeddings.embed_query("what is the second letter of greek alphabet")
```
